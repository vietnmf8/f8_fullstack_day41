```python
# Nỗi đau khi chưa có RTK Query
## Nỗi đau số 1:  Code lặp đi lặp lại:
- Cách cũ: (Dùng `useEffect` hoặc `createAsyncThunk`): Để tải danh sách sản phẩm, phải làm thủ công 3 bước:
    1. Khái báo state: `loading`, `data`, `error`
    2. Khi bắt đầu gọi API: set `loading = true`
    3. Khi gọi API thành công: set `loading = false`, lưu `data`
    4. Khi gọi API thất bại: set `loading = false`, lưu `error`

=> Phải viết logic này lặp lại cho mọi API (Sản phẩm, bài viết, user) --> Code sẽ phình to

## Giải pháp RTK Query:
- Tự động làm hết việc đó
- Chỉ cần gọi Hook --> nó trả về luôn 3 biến
    
```

```python
## Nỗi đau số 2: Quản lý Cache (Bộ nhớ đệm)
- Cách cũ:
    1. Bạn vào trang "Sản phẩm" -> gọi API lấy danh sách sản phẩm
    2. Bạn chuyển sang trang "Giới thiệu" -> rồi quay lại trang "Sản phẩm"
- Vấn đề:
    - Ứng dụng sẽ gọi API lại từ đầu, người dùng lại phải nhìn thấy Loading.
    - Dù dữ liệu chưa hề thay đổi. 
    - Muốn sửa, bạn phải tự viết logic kiểm tra: "Dữ liệu này đã tải chưa? Nếu có rồi thì đừng gọi API nữa"
- Giải pháp: Nó có sẵn bộ nhớ đệm (Cache)
    - Lần 1 vào trang: Gọi API -> Lưu vào Cache
    - Lần 2 quay lại: Lấy ngay từ Cache ra hiển thị (không gọi API, không hiện Loading) => Trải nghiệm UX tốt
=> Giải quyết: Tối ưu hiệu năng, giảm tải cho Server, trải nghiệm mượt mà
```

```python
## Nỗi đau số 3: Dữ liệu bị cũ (Stale Data) & Đồng bộ 
- Cách cũ:
    - Bạn có danh sách 10 sản phẩm
    - Bạn bấm nút "Xoá" sản phẩm 1 -> API xoá thành công trên Server
- Vấn đề:
    - Trên giao diện (Client) danh sách vẫn còn 10 cái (Vì Redux state chưa cập nhật)
- Hệ quả:
    - Phải tự viết thêm logic: "Xoá xong thì tìm trong mảng state cũ, lọc bỏ cái item đó đi"
    - Hoặc: "Xoá xong thì gọi lại API lấy danh sách mới"
- Giải pháp: RTK Query (Tính năng Tags)
    - Bạn chỉ cần bảo: "Hành động Xoá này làm thay đổi danh sách Products"
    - RTK Query tự động hiểu: "À, danh sách Products cũ rồi, mình phải tự gọi lại API lấy cái mới nhất về"

=> Giải quyết: Không bao giờ lo dữ liệu hiển thị sai lệch so với Server
```

```python
# TÓM TẮT: RTK Query là gì?

- RTK Query là một thư viện đi kèm sẵn trong Redux Toolkit dùng để:
    1. Fetching: Gọi API lấy dữ liệu
    2. Caching: Lưu trữ thông minh, tránh gọi thừa
    3. Synchronization: Tự động đồng bộ dữ liệu giữa Server và Client
```
