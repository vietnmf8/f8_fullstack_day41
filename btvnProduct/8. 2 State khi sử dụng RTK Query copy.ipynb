```python
- Hãy tưởng tượng toàn bộ ứng dụng là một VĂN PHÒNG CÔNG TY
# State 1: productApi (RTK Query) = Kho hàng Tiki (Server State)
    - Đặc điểm: Là nơi chứa dữ liệu "gốc" lấy từ nhà cung cấp (Server) về (dữ liệu từ API phía Backend trả về)
    - Cơ chế: 
        - Bạn không thể tự ý chạy vào nhà kho để sắp xếp lại hàng hoá hay dán lại tem giá (Không tự ý sửa được API)
        - Bạn chỉ có thể YÊU CẦU (Request API) (Query - GET / Mutation (Thay đổi - Thêm/Sửa/Xoá)) để nhân viên kho (RTK Query) tự đi lấy hàng, tự cập nhật, tự báo hàng về hay chưa
    - Vai trò:
        - Đồng bộ dữ liệu với Server
        - Lo việc Caching (lưu kho)
        - Lo việc Loading (vận chuyển)

# State 2: productSlice (Slice thường) = Cái bàn làm việc của bạn (Client/UI State)
    - Đặc điểm: Là nơi chứa những thứ chỉ tồn tại trên máy tính của bạn lúc này, không liên quan đến Server
    - Cơ chế: Toàn quyền quyết định. 
        + Bạn muốn đặt cái bút ở đâu, 
        + Muốn mở cửa cuốn sổ nào ra (Bật/Tắt Modal), 
        + Muốn đánh dấu trang nào (Selected Item)
    - Vai trò:
        - Quản lý giao diện người dùng (UI), các trạng thái TẠM THỜI
```
