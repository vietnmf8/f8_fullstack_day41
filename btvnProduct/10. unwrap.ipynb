.unwrap()

- RTK Query gốc: Luôn trả về `{ data, error }`. Không bao giờ throw error. -> Khó dùng try/catch.
    - Mặc định, RTK Query được thiết kế để không bao giờ làm sập ứng dụng (crash-safe). Khi API trả về lỗi 500 (Server Error), RTK Query không ném lỗi. Thay vào đó, nó âm thầm đóng gói lỗi đó vào một object và trả về nhẹ nhàng:
        1. Thành công: Trả về `{ data: {id: 1, ...} }`
        2. Thất bại: Trả về `{ error: {status: 500, message: "..."} }`
    => Cả hai trường hợp trên đều là Promise được giải quyết thành công (Resolved).
        
- `.unwrap()`: Trả về `data` HOẶC `throw error`. -> Dễ dùng try/catch.
